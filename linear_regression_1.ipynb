Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Creating a random line with bias for every single datom.

```python
a = np.random.random(1) * 10
b = np.random.random(1) * 10
X = np.random.random(100) * 10
bias = np.random.random(100) - 0.5
y = a * (X + bias) + b
```

Plotting the data:

```python
plt.scatter(X, y)
```

setting initial a_ and b_ and then changing them to reach closer values to starting values a and b:

```python
a_ = 0
b_ = 0
alpha = 0.001 / len(X)
for _ in range(1_000_000):
    b_ -= alpha * np.sum(a_ * X + b_ - y)
    a_ -= alpha * np.sum((a_ * X + b_ - y) * X)
print(a, a_)
print(b, b_)
```

Here I visualized the line:

```python
plt.scatter(X, y)
X_ = np.linspace(min(X), max(X), 10)
plt.plot(X_, a_ * X_ + b_, c='red')
```

Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Creating random data with bias:

```python
a = np.random.random(1) * 10
b = np.random.random(1) * 10
X = np.random.random(100) * 10
bias = np.random.random(100) - 0.5
y = a * (X) + b + bias
```

plotting the data:

```python
plt.scatter(X, y)
```

calculating the line mathematically:

```python
x_mean = np.mean(X)
y_mean = np.mean(y)
n = len(X)
a_1 = (n * x_mean * y_mean - np.sum(X * y)) / (n * x_mean ** 2 - np.sum(X ** 2))
b_1 = y_mean - a_1 * x_mean
```

plotting the line fit to data:

```python
plt.scatter(X, y)
plt.plot(X, a_1 * X + b_1, c='red')
```

Using gradient descent to compare results:

```python
a_2 = 0
b_2 = 0
alpha = 0.001 / len(X)
for _ in range(10_000):
    b_2 -= alpha * np.sum(a_2 * X + b_2 - y)
    a_2 -= alpha * np.sum((a_2 * X + b_2 - y) * X)
print('initial:', a, 'gradient descent:', a_1, 'mathematical:', a_2)
print('initial:', b, 'gradient descent:', b_1, 'mathematical:', b_2)
```

plotting result using gradient descent:

```python
plt.scatter(X, y)
plt.plot(X, a_2 * X + b_2, linestyle='dashed', c='orange')
```

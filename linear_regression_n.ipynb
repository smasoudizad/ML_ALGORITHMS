Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Creating a random polynomial with bias for every single datom:

```python
n = 6
w = (np.random.random(n + 1) - 0.5) * 100
X = np.random.random(100)
bias = (np.random.random(100) - 0.5)

def f(X, w):
    y = np.zeros(len(X))
    for i in range(len(w)):
        y += w[i] * X ** i
    return y

y = f(X, w) + bias
```

Plotting the data:

```python
plt.scatter(X, y)
```

Predicting using gradient descent:

```python
w_ = np.zeros(n + 1)
alpha = 0.01 / len(X)
for _ in range(100_000):
    c = 1
    for i in range(n + 1):
        w_[i] -= alpha * np.sum((f(X, w_) - y) * X ** i)

print('Real:', w)
print('Predicted:', w_)
```

```python
plt.scatter(X, y)
X_ = np.linspace(min(X), max(X), 100_000)
y_ = 0
for i in range(n + 1):
    y_ += w[i] * X_ ** i
plt.plot(X_, y_, c='red')
```
